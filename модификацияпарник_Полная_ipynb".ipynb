```python
import pandas as pd

# Загрузка CSV-файла с указанием кодировки, разделителя и пропуском проблемных строк
df = pd.read_csv('Теплица 14 (слева)_Состояние_2023.csv', encoding='windows-1251', sep=';', on_bad_lines='skip')

# Вывод первых строк для проверки
print(df.head())
# Вывод названий всех столбцов
print(df.columns)

# После этого шага, убедитесь, что столбец "Направление ветра" (или как он там называется) действительно существует, и затем продолжим.
```

```python
# Извлечение числовой части из столбца "Направление ветра"
df['Направление ветра'] = df['Направление ветра'].str.extract('(\d+)').astype(int)
# Удаление столбцов 'Ключ' и 'Время суток'
df.drop(['Обработка Состояние', 'Давление воды обратной', 'СИО Насос Состояние','Авария Состояние'], axis=1, inplace=True)

# Сохранение изменений в новый CSV-файл
df.to_csv('Теплица 14 (слева)_Состояние_2023_modified.csv', index=False, sep=';')

# Показать результат
print(df.head())

# Сохранение изменений в новый CSV-файл
df.to_csv('Теплица 14 (слева)_Состояние_2023_modified.csv', index=False, sep=';')

# Показать результат
print(df.head())
```

```python
# Установка ipywidgets, если еще не установлено
!pip install ipywidgets

import pandas as pd
import ipywidgets as widgets
from IPython.display import display

# Загрузка данных
df = pd.read_csv('Теплица 14 (слева)_Состояние_2023_modified.csv', sep=';')

# Получение списка всех столбцов
all_columns = df.columns.tolist()

# Создание виджета для множественного выбора
multi_select = widgets.SelectMultiple(
    options=all_columns,
    description='Столбцы',
    disabled=False
)

# Кнопка для применения выбора
apply_button = widgets.Button(description="Применить")

# Функция для обработки нажатия кнопки
def on_apply_button_clicked(b):
    selected_columns = list(multi_select.value)
    new_df = df[selected_columns]
    print("Оставлены следующие столбцы:")
    print(new_df.head())
    new_df.to_csv('Теплица_оставлены_избранные_столбцы.csv', index=False, sep=';')
    print("Новый файл сохранен как 'Теплица_оставлены_избранные_столбцы.csv'")

# Привязка функции к кнопке
apply_button.on_click(on_apply_button_clicked)

# Отображение виджетов
display(multi_select)
display(apply_button)
```

```python
import pandas as pd
import numpy as np

# Загрузка CSV-файла
df = pd.read_csv('Теплица_оставлены_избранные_столбцы.csv', encoding='windows-1251', sep=';', error_bad_lines=False)

# Замена запятых на точки и преобразование в числовой формат для всех столбцов
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = df[col].str.replace(',', '.')
        try:
            df[col] = df[col].astype(float)
        except ValueError:
            pass

# Усреднение до ближайшей половины единицы для всех числовых столбцов
for col in df.select_dtypes(include=[np.number]).columns:
    df[col] = np.round(df[col] * 2) / 2

# Сохранение изменений в новый CSV-файл
df.to_csv('Теплица_усредненные_данные.csv', index=False, encoding='windows-1251',sep=';')
```
